# 2. The Question: Decision, Action, Cost of a Wrong Call

### Research Question

Can observable search performance and content signals be used to identify and rank pages that should be reviewed first for potential content refresh?

### Decision

The decision is which pages should be reviewed before others.

### Action

An SEO specialist or content manager can review the highest-ranked pages first and decide whether they need to be updated, optimized, or investigated further.

### Cost of a Wrong Recommendation

If the model ranks pages incorrectly, valuable pages that need attention may be missed, while time is spent reviewing pages that are already performing well. This reduces the efficiency of the content review process.

### Why Machine Learning?

Instead of relying on a single rule, machine learning can combine multiple observable signals such as impressions, CTR, content age, search position, and word count to support better prioritization decisions.

```python
# ============================================
# Quick Look at the Data
# ============================================

import os
import subprocess
import sys
import pandas as pd

REPO_DIR = "flyrank-ml-internship"
REPO_URL = "https://github.com/MtalhaShzd/flyrank-ml-internship.git"

# Clone your repository if it doesn't already exist
if not os.path.isdir(f"/content/{REPO_DIR}"):
    subprocess.run(
        ["git", "clone", REPO_URL],
        check=True
    )

# Change to repository root
os.chdir(f"/content/{REPO_DIR}")

print("Current Working Directory:", os.getcwd())

# Load the dataset
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("=" * 50)
print("Dataset Overview")
print("=" * 50)

print(f"Total Pages   : {len(df)}")
print(f"Total Columns : {len(df.columns)}")

# Discovery 1
corr_search = df["search_volume"].corr(df["impressions_90d"])

# Discovery 2
corr_position = df["avg_position"].corr(df["ctr"])

# Discovery 3
median_age = df.groupby("trend_direction")["content_age_days"].median()

print("\nCorrelation between search_volume and impressions_90d:")
print(round(corr_search, 3))

print("\nCorrelation between avg_position and CTR:")
print(round(corr_position, 3))

print("\nMedian Content Age by Trend Direction:")
print(median_age)
```

# 3. Quick Look at the Data

The starter dataset contains **30,000 pages**, providing a realistic dataset for experimentation.

The correlation between **search volume** and **impressions** is approximately **0.001**, indicating that keyword search volume alone does not strongly explain page impressions.

The correlation between **average position** and **CTR** is approximately **-0.073**, showing only a very weak relationship.

The median content age also differs across trend groups. Stable pages have a higher median content age (300 days), while down pages have a lower median content age (216 days). These observations suggest that multiple signals may be useful when deciding which pages should be reviewed instead of relying on a single metric.
